```python
import numpy as np
import copy
import matplotlib.pyplot as plt

# a sensor
def sense(x):
    return x
```

```python
def simulate(Δt, x, u):
    # The control input is equal to the velocity of the robot
    dx = np.array(u)
    # Euler integration
    x += Δt * u
    return x
```

```python
# Question 1
def control1(t, y):
    ### WRITE YOUR CONTROL POLICY HERE:
    ux = -4*(np.sin(t))
    uy = 2*(np.cos(t))
    return np.array([ux, uy])
```

```python
tf = 10 # Final time
Δt = 0.1    # Time step
time = np.linspace(0.,tf, int(tf / Δt) + 1)  # Time interval


# Initial conditions
x = np.array([7.2, 2])
x_log = [copy.copy(x)]

for t in time:
    y = sense(x)
    u = control1(t, y)    
    x = simulate(Δt, x, u)
    x_log.append(copy.copy(x)) # Used to recorded all of points passed
    
x_log = np.array(x_log)
# Plot the pictures
plt.plot(x_log[:,0], x_log[:,1],'r.')
```

```python
# Question 3
# : Rotated eight curve
def control2(t, y):
    ### WRITE YOUR CONTROL POLICY HERE:
    ux = np.sin(t) 
    uy = -(np.sin(t) * np.sin(t)) + (np.cos(t) * np.cos(t))
    return np.array([ux, uy])
```

```python
tf = 2*np.pi # Final time
Δt = 0.1    # Time step
time = np.linspace(0.,tf, int(tf / Δt) + 1)  # Time interval

# Initial conditions
x = np.array([-1., 0.])
x_log = [copy.copy(x)]

for t in time:
    y = sense(x)
    u = control2(t, y)    
    x = simulate(Δt, x, u)
    x_log.append(copy.copy(x)) # Used to recorded all of points passed
    
x_log = np.array(x_log)
# Plot the pictures
plt.plot(x_log[:,0], x_log[:,1],'.r')
```

```python
def simulate2(Δt, x, u):
    # The control input is equal to the velocity of the robot
    dx = np.array(u)
    
    # random wind
    #wind = np.random.normal(0, 0.1, 2)  # Mean 0, std deviation 0.1, for x and y
    # constant wind
    wind = np.array([0.1,0.1])
    # Euler integration
    x += Δt * (dx  + wind)
    return x 
```

```python
def control3(t, y):
    ### WRITE YOUR CONTROL POLICY HERE:
    ux = -np.sin(t)
    uy = np.cos(t)
    return np.array([ux, uy])
```

```python
def simulate3(Δt, x, u):
    # The control input is equal to the velocity of the robot
    dx = np.array(u)
    
    # random wind
    wind = np.random.normal(0, 0.1, 2)  # Mean 0, std deviation 0.1, for x and y
    # constant wind
    #wind = np.array([0.1,0.1])
    # Euler integration
    x += Δt * (dx  + wind)
    return x 
```

```python
tf =70  # Final time
Δt = 0.1    # Time step
time = np.linspace(0.,tf, int(tf / Δt) + 1)  # Time interval


# Initial conditions
x = np.array([0., 0.])
x_log = [copy.copy(x)]

for t in time:
    y = sense(x)
    u = control3(t, y)
    x = simulate2(Δt, x, u) 
    x_log.append(copy.copy(x))
    
x_log = np.array(x_log)
# Plot the pictures
plt.plot(x_log[:,0], x_log[:,1],'.r')
```
